```python
from contrasttransferfunction.spectrumhelpers import downscale_spectrum, adjust_central_cross, subtract_baseline, cosine_highpass, radial_average
import matplotlib.pyplot as plt
import numpy as np
```

```python
spectrum = np.load("../tests/test_data/spectrum.npy")
plt.imshow(spectrum, cmap="gray")
spectrum = spectrum.astype(np.float32)
print(spectrum.max())
```

```python
s2 = downscale_spectrum(spectrum, 512)
plt.imshow(s2,cmap="gray")
print(s2.max())
```

```python
a = adjust_central_cross(s2)
```

```python
plt.imshow(a,cmap="gray")
```

```python
b = subtract_baseline(a)
```

```python
plt.imshow(b,cmap="gray")
```

```python
plt.imshow(a-b,cmap="gray")
```

```python
c= cosine_highpass(a-b,pixel_size_angstroms=4.24)
```

```python
plt.imshow((a-b)*c,cmap="gray")
```

```python
plt.plot(radial_average((a-b)*c))
```
